```r
df <- read.csv("https://raw.githubusercontent.com/LesTa98/northstar-data-analysis/refs/heads/main/data/cleaned/cleaned_data.csv")
```

```r
install.packages("sqldf")
install.packages("tidyverse")
install.packages("lubridate")

library(sqldf)
library(tidyverse)
library(lubridate)
```

SQL QUERIES

1. Average delay by pickup zone

Shows which pickup zones experience the longest delivery delays on average.

```r
sqldf("SELECT pickup_zone, AVG(delay_minutes) AS avg_delay
       FROM df
       GROUP BY pickup_zone
       ORDER BY avg_delay DESC")
```

2. Average delay by hub

Identifies hubs with the highest average delivery delays.

```r
sqldf("SELECT hub_name, AVG(delay_minutes) AS avg_delay
       FROM df
       GROUP BY hub_name
       ORDER BY avg_delay DESC")
```

3. Worst 10 drivers by average delay

```r
sqldf("SELECT driver_id, AVG(delay_minutes) AS avg_delay
       FROM df
       GROUP BY driver_id
       ORDER BY avg_delay DESC
       LIMIT 10")
```

4. Top 10 customers by loyalty

```r
sqldf("SELECT customer_id, MAX(loyalty_score) AS loyalty_score
       FROM df
       GROUP BY customer_id
       ORDER BY loyalty_score DESC
       LIMIT 10;")
```

5. Worst 10 vehicles by average delay

```r
sqldf("SELECT vehicle_id, AVG(delay_minutes) AS avg_delay
       FROM df
       GROUP BY vehicle_id
       ORDER BY avg_delay DESC
       LIMIT 10")
```

6. Worst customers by average delay

```r
sqldf("SELECT customer_id, AVG(delay_minutes) AS avg_delay
       FROM df
       GROUP BY customer_id
       ORDER BY avg_delay DESC
       LIMIT 10")
```

7. Delay trend by Month

```r
sqldf("SELECT strftime('%m', dispatch_time) AS month,
       AVG(delay_minutes) AS avg_delay
       FROM df
       GROUP BY month
       ORDER BY month")
```

8. Delay trend by Weekday

```r
sqldf("SELECT strftime('%w', dispatch_time) AS weekday,
       AVG(delay_minutes) AS avg_delay
       FROM df
       GROUP BY weekday
       ORDER BY weekday")
```

9. Complaints vs no complaints

```r
sqldf("SELECT has_complaint, AVG(delay_minutes) AS avg_delay
       FROM df
       GROUP BY has_complaint")
```

10. Incidents vs no incidents

```r
sqldf("SELECT has_incident, AVG(delay_minutes) AS avg_delay
       FROM df
       GROUP BY has_incident")
```

```r
11. High delay delivery counts per hub
```

```r
sqldf("SELECT hub_name, COUNT(*) AS high_delay_count
       FROM df
       WHERE delay_minutes > 60
       GROUP BY hub_name
       ORDER BY high_delay_count DESC")
```
